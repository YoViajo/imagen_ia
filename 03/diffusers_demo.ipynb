Hugging Face Just Released the Diffusers Library

Inicializar modelo o pipeline

```python
from diffusers import DiffusionPipeline
```

```python
model_id = "CompVis/ldm-text2im-large-256"

# load text input to diffusion pipeline
ldm = DiffusionPipeline.from_pretrained(model_id)
```

Crear una entrada y ejecutarla en el pipeline *ldm*

Generar una pintura de una ardilla comiendo un plátano

```python
# run inference (sample random noise and denoise)
prompt = "A painting of a squirrel eating a banana"
images = ldm([prompt], num_inference_steps=50, eta=.3, guidance_scale=6)
```

```python
images['sample'][0]
```

Otra pintura...

```python
images = ldm(
    [prompt],
    num_inference_steps=100,
    eta=.3,
    guidance_scale=6
)
images['sample'][0]
```
